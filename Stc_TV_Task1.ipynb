```python
# Here we try to get the most watched movies (Total Views / Total Users Views / Total watch time)
# For series we concatenated the Session episode to differentiate between episodes
grouped=df.copy()
grouped.loc[grouped['program_class'] == 'SERIES/EPISODES', 'program_name'] = grouped['program_name']+'_SE'+grouped['season'].astype(str)+'_EP'+grouped['episode'].astype(str)
grouped = grouped.groupby(['program_name','program_class'])\
.agg({'user_id_maped': [('co1', 'nunique'),('co2', 'count')],\
      'duration_seconds': [('co3', 'sum')] }).reset_index()
grouped.columns = ['program_name','program_class','No of Users who Watched', 'No of watches', 'Total watch time in seconds']
grouped['Total watch time in houres']=grouped['Total watch time in seconds']/3600
grouped = grouped.drop(columns=['Total watch time in seconds'])
grouped = grouped.sort_values(by=['Total watch time in houres', 'No of watches','No of Users who Watched'], ascending=False).reset_index(drop=True)
```


## 🕒Temporal analysis

```python
# تحويل عمود التاريخ إلى datetime

df['date_'] = pd.to_datetime(df['date_'], errors='coerce')
df['day_of_week'] = df['date_'].dt.day_name()
df['month'] = df['date_'].dt.month_name()
```

```python
# Views by day of the week
import seaborn as sns
sns.set_theme(style="darkgrid")

plt.figure(figsize=(8, 5))
order_days = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
sns.countplot(data=df, x='day_of_week', order=order_days, color='#a0bcd0')
plt.title('View Count by Day of the Week')
plt.xlabel('Day')
plt.ylabel('Number of Views')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pie chart - views by day of week with custom colors
day_counts = df['day_of_week'].value_counts().reindex(order_days)

day_colors = ['#8cb3d9', '#a0c4ff', '#b9d7ea', '#c6def1', '#d0e8f2', '#e3f2fd', '#ccdff7']

plt.figure(figsize=(6, 6))
plt.pie(day_counts, labels=day_counts.index, autopct='%1.1f%%', startangle=140, colors=day_colors)
plt.title('View Distribution by Day of the Week')
plt.axis('equal')
plt.show()
```

```python
# Views by Month
plt.figure(figsize=(10, 5))
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
sns.countplot(data=df, x='month', order=month_order, color='#b5d6c3')
plt.title('View Count by Month')
plt.xlabel('Month')
plt.ylabel('Number of Views')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pie chart - views by month with custom colors
month_counts = df['month'].value_counts().reindex(month_order)

month_colors = ['#d4f1be', '#b2e7b6', '#a0d6a1', '#8fcb9b', '#7ecf9f', '#6dc8a2',
                '#5cbfa3', '#4db5a7', '#3eada8', '#2fa6aa', '#1e9fad', '#0e97af']

plt.figure(figsize=(8, 8))
plt.pie(month_counts, labels=month_counts.index, autopct='%1.1f%%', startangle=140, colors=month_colors)
plt.title('View Distribution by Month')
plt.axis('equal')
plt.show()
```
